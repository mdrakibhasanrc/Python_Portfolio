# Data Preprocessing

```python
#extract year from Date Colimns
df['hire_year']=df['hire_date'].astype('str').str.split('-').str[0]
```

```python
#extract Month from Date Colimns
df['hire_Month']=df['hire_date'].astype('str').str.split('-').str[1]
```

```python
df['hire_month_name']=df['hire_date'].dt.month_name()
```

```python
df.drop(columns=['hire_Month'],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['hire_year']=df['hire_year'].astype('int')
df['hire_month_name']=df['hire_month_name'].astype('object')
```

```python
#extract Month from Date Colimns
df['termnation_year']=df['term_date'].astype('str').str.split('-').str[0]
```

```python
df.drop(columns=['term_date'],axis=1,inplace=True)
```

```python
df.drop(columns=['hire_date'],axis=1,inplace=True)
```
